````python
# 2. Group the data by 'Year' to get the total sales for the line plot.
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum()

# 3. Create the base plot.
plt.figure(figsize=(12, 7))
yearly_sales.plot(kind='line', marker='o')

# 4. Set the title.
plt.title('Automobile Sales during Recession')

# 5. Set the x-axis ticks to show all years and rotate the labels for readability.
plt.xticks(yearly_sales.index, rotation=90)

# 6. Add annotations for the recession years.
# Get a list of unique years that had a recession from the loaded data.
recession_years = df[df['Recession'] == 1]['Year'].unique()

# Loop through each recession year to add an annotation.
for year in recession_years:
    # Get the corresponding sales value for that year.
    sales_value = yearly_sales.loc[year]

    # Add the annotation.
    plt.annotate('Recession',
                 xy=(year, sales_value),
                 xytext=(year, sales_value + 2000), # Adjust text position
                 arrowprops=dict(facecolor='black', shrink=0.05),
                 ha='center',
                 fontsize=9)

# 7. Add labels and a grid for clarity.
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# 8. Save and display the final plot.
plt.savefig('Line_Plot_1.png')
plt.show()
````

### Now you want to compare the sales of different vehicle types during a recession and a non-recession period

````python
# 1. Create a new column to label the periods as 'Recession' or 'Non-Recession'.
df['Period'] = df['Recession'].apply(lambda x: 'Recession' if x == 1 else 'Non-Recession')

# 2. Group the data by 'Period' and 'Vehicle_Type' and calculate the AVERAGE sales.
avg_sales_comparison = df.groupby(['Period', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# 3. Create a bar chart to compare the average sales.
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_sales_comparison,
            x='Vehicle_Type',
            y='Automobile_Sales',
            hue='Period',
            palette={'Recession': '#ff6361', 'Non-Recession': '#58508d'})

# 4. Add a title and labels for clarity.
plt.title('Average Automobile Sales: Recession vs. Non-Recession', fontsize=16)
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.tight_layout()

# 5. Save and display the plot.
plt.savefig('Bar_Chart.png')
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
# 1. Create separate DataFrames for recession and non-recession periods.
rec_data = df[df['Recession'] == 1]
non_rec_data = df[df['Recession'] == 0]

# 2. Create a figure that will contain the subplots.
fig = plt.figure(figsize=(12, 6))

# 3. Create the first subplot for the recession period.
ax0 = fig.add_subplot(1, 2, 1) # (1 row, 2 columns, first plot)
sns.lineplot(x='Year', y='GDP', data=rec_data, label='Recession', ax=ax0)
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')
ax0.set_title('GDP Variation during Recession Period')
ax0.grid(True, linestyle='--', alpha=0.6)

# 4. Create the second subplot for the non-recession period.
ax1 = fig.add_subplot(1, 2, 2) # (1 row, 2 columns, second plot)
sns.lineplot(x='Year', y='GDP', data=non_rec_data, label='Non-Recession', ax=ax1)
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.set_title('GDP Variation during Non-Recession Period')
ax1.grid(True, linestyle='--', alpha=0.6)

# 5. Adjust the layout to prevent titles/labels from overlapping.
plt.tight_layout()

# 6. Save and display the figure.
plt.savefig('Subplot.png')
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
# 1. Filter the DataFrame to include only data from non-recession periods.
non_rec_data = df[df['Recession'] == 0]

# 2. Extract the 'Seasonality_Weight' to use for the bubble sizes.
size = non_rec_data['Seasonality_Weight']

# 3. Create the bubble plot using seaborn's scatterplot function.
plt.figure(figsize=(12, 7))
sns.scatterplot(data=non_rec_data, x='Month', y='Automobile_Sales', size=size, hue='Seasonality_Weight', legend=False, sizes=(30, 1200))

# 4. Add a title and labels for clarity.
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality Impact on Automobile Sales')

# 5. Customize the x-axis to show month names instead of numbers.
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(ticks=range(1, 13), labels=month_labels)

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# 6. Save and display the plot.
plt.savefig('Bubble.png')
plt.show()
````

 ### TASK 1.7: Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods.
 <br>How did the advertising expenditure of XYZAutomotives change during recession and non-recession periods? 


<details><summary>Click here for a hint</summary>

<p>
     You can create two dataframe for recession and nonreccession period.
    <br> Calculate the sum of Advertising_Expenditure for both dataframes
    <br> Pass these total values to plt.pie(). May include labels as ['Recession', 'Non-Recession']
        <br>Feel Free to customie the pie further
    <br>title this plot as  - Advertising Expenditure during Recession and Non-Recession Periods</p>
</details>


````python
# 1. Create separate DataFrames for recession and non-recession periods.
rec_data = df[df['Recession'] == 1]
non_rec_data = df[df['Recession'] == 0]

# 2. Calculate the total advertising expenditure for each period.
rec_expenditure = rec_data['Advertising_Expenditure'].sum()
non_rec_expenditure = non_rec_data['Advertising_Expenditure'].sum()

# 3. Create the pie chart.
plt.figure(figsize=(8, 8))
labels = ['Recession', 'Non-Recession']
sizes = [rec_expenditure, non_rec_expenditure]
colors = ['#ff6361', '#58508d']
explode = (0.1, 0)  # "Explode" the recession slice for emphasis

plt.pie(sizes,
        explode=explode,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%', # Adds percentage labels to the slices
        shadow=True,
        startangle=140)

# 4. Add a title.
plt.title('Advertising Expenditure during Recession and Non-Recession Periods', fontsize=16)
plt.axis('equal')  # Ensures the pie chart is a perfect circle.

# 5. Save and display the plot.
plt.savefig('Pie_1.png.png')
plt.show()
````
